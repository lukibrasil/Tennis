First Thing I coded here was the algorithim to find the nearest line that the ball landed in. This was incorporated with the hawkeye dataset

```python
# This function calculates the distance of a point from the different lines in a tennis court
def find_nearest_line(coord):
    # Split the given coordinate into x and y
    x, y = coord

    # Dictionary that stores the distances from each relevant line on the tennis court
    distances = {
        "x=-11.8": abs(x + 11.8),  # Distance from left sideline
        "x=11.8": abs(x - 11.8),  # Distance from right sideline
        "y=4.115": abs(y - 4.115),  # Distance from bottom baseline
        "y=-4.115": abs(y + 4.115),  # Distance from top baseline
        "x=6.4": abs(x + 6.4),  # Distance from left service line
        "x=-6.4": abs(x - 6.4),  # Distance from right service line
        "x=0": abs(x),  # Distance from center line (vertical)
        "y=0": abs(y)  # Distance from center line (horizontal)
    }

    # Find the line with the minimum distance to the coordinate
    closest_line = min(distances, key=distances.get)
    
    # Return the line that's closest to the coordinate
    return closest_line

# This function maps the nearest line to a court acronym
def find_nearest_acronym(coord):
    x, y = coord

    # Use previous function to find the closest line
    closest_line = find_nearest_line(coord)

    # Map the closest line to the respective court acronym
    if closest_line in ["y=4.115", "y=-4.115"]:
        acronym = "F"  # 'F' for Forecourt
    elif closest_line in ["x=-11.8", "x=11.8"]:
        acronym = "B"  # 'B' for Backcourt
    elif closest_line in ["x=-6.4", "x=6.4"]:
        acronym = "S"  # 'S' for Service court
    elif closest_line == "x=0":
        acronym = "N"  # 'N' for Net
    elif closest_line == "y=0" and -11.8 < x < 11.8:
        acronym = "C"  # 'C' for Center
    else:
        # if the x coordinate is outside the range of the court, then it's Backcourt
        # otherwise it's Center
        acronym = "B" if x < -11.8 or x > 11.8 else "C"
    
    # Return the court acronym for the closest line
    return acronym

# This function combines the previous functions and adds a side designation
def find_label(coord):
    x, y = coord

    # Find the closest line using the previously defined function
    closest_line = find_nearest_line(coord)

    # If the closest line is the center horizontal line and x is within the range of the court
    # then it's Center, otherwise use the previous function to find the acronym
    if closest_line == "y=0" and -11.8 < x < 11.8:
        acronym = "C"
    else:
        acronym = find_nearest_acronym(coord)

    # Determine the side based on the x coordinate
    if x == 0:  # Directly in the center
        side = ""
    elif x < 0:  # Right side of the court from the server's perspective
        side = "R"
    else:  # Left side of the court from the server's perspective
        side = "L"

    # Return the combined label with side and acronym
    return side + acronym

# Apply the 'find_label' function to each row in the hawkeye DataFrame
# The function is applied to the 'x1' and 'y1' columns and the result is stored in a new column called 'label'
hawkeye['label'] = hawkeye.apply(lambda row: find_label((row['x1'], row['y1'])), axis=1)
```

```python
def tennis_points(value):
    value = int(value)
    if value == 0:
        return 0
    elif value == 1:
        return 15
    elif value == 2:
        return 30
    elif value == 3:
        return 40
    else:
        return 50

# Split the 'score' column into two columns
hawkeye[['server_score', 'receiver_score']] = hawkeye['score'].str.split(' ', expand=True)

# Convert the tennis points to their actual values
hawkeye[['server_score', 'receiver_score']] = hawkeye[['server_score', 'receiver_score']].applymap(tennis_points)
```
